```python
#Kullanacağımız kitaplıkları içe aktaralım
import csv
import datetime
import pandas as pd

#Main fonksiyon ile oluşturduğumuz Menu.txt dosyasını okuyalım
Menu = open("Menu.txt", "r")
oku = Menu.read()

def main() :
    print(oku)

if __name__ == "__main__":
    main()

#csv dosyamıza kaydetmek üzere zamanı, kullanıcı ismini, tc, kart bilgilerini alalım
siparis_zamani = datetime.datetime.now()
tarih = datetime.datetime.ctime(siparis_zamani)
print("Sipariş Zamanı: ", tarih)
isim = input("İsminiz: ")
tc = input("TC Kimlik Numaranız: ")
kart = input("Kredi Kartı Numaranız: ")
sifre = input("Kredi Kartı Şifreniz: ")

#Pizzalarımızı tanımlayalım
description_mappings = {
    1: "Klasik",
    2: "Margarita",
    3: "Türk Pizza",
    4: "Sade Pizza"
}

#Fiyatları belirleyelim
cost_mappings = {
    "Klasik": 90,
    "Margarita": 100,
    "Türk Pizza": 120,
    "Sade Pizza": 80
}

#Classlarımızı oluşturalım
class Pizza():
    def __init__(self, description):
        self.description = description

    def set_description(self, description):
        self.description = description

    def get_description(self):
        return self.description()

    def get_cost(self):
        return cost_mappings[self.description]

class Order():
    def __init__(self):
        self.pizzas = []

    def addPizza(self, pizza):
        self.pizzas.append(pizza)

    def getTotal(self):
        total = 0
        for pizza in self.pizzas:
            total += pizza.get_cost()
        return total

order = Order()

#Kullanıcıdan siparişini almaya başlayalım 
def run():
    print("\nHangi pizzayı seçmek istersiniz?\n\n\
    _____________________________________________________________________\n\
    | 1: Klasik   |  2: Margarita  |  3: Türk Pizza    |  4: Sade Pizza |\n\
    |     $90     |     $100       |        $120       |       $80      |\n\
    |_____________|________________|___________________|________________|\n\
    \n- Seçiminizi onaylamak için t tuşuna basınız.\n")

    while True:
        try:
            
            response = input('-')

            if response == 't':
                break

            pizza_wanted = int(response)

            pizza_wanted = description_mappings[pizza_wanted]
            

            print(f"Pizza: {pizza_wanted}")
            if int(response) == 1:                              #Seçilen siparişi seçilen numaraya göre csv dosyamıza kaydedelim
                with open("Orders_Database.csv", "a", encoding='utf-8') as f:
                  yazıcı = csv.writer(f)
                  yazıcı.writerow(['Klasik', tarih, isim, tc, kart, sifre])
            elif int(response) == 2:
                with open("Orders_Database.csv", "a", encoding='utf-8') as f:
                  yazıcı = csv.writer(f)
                  yazıcı.writerow(['Margarita', tarih, isim, tc, kart, sifre])  
            elif int(response) == 3:
              with open("Orders_Database.csv", "a", encoding='utf-8') as f:
                yazıcı = csv.writer(f)
                yazıcı.writerow(['Türk Pizza', tarih, isim, tc, kart, sifre])
            elif int(response) == 4:
              with open("Orders_Database.csv", "a", encoding='utf-8') as f:
                yazıcı = csv.writer(f)
                yazıcı.writerow(['Sade Pizza', tarih, isim, tc, kart, sifre])
            

            order.addPizza(Pizza(pizza_wanted))
        except:
            print("Yanlış seçim yaptınız, lütfen tekrar deneyin.")
   
run()

print("Sipariş tutarı: ", "$" + str(order.getTotal()))

#Sipariş seçim ve tutar hesabı bittikten sonra pizzanın üzeri için seçilecek ek malzemeleri tanımlayalım
topping_mappings = {
                    1: 'Zeytin', 
                    2: 'Mantar', 
                    3: 'Keçi Peyniri', 
                    4: 'Et', 
                    5: 'Soğan', 
                    6: 'Mısır', 
                    }

#Ek malzemelerin fiyatını belirleyelim
topping_cost_mappings = {
                        'Zeytin': 2, 
                        'Mantar': 3, 
                        'Keçi Peyniri': 5, 
                        'Et': 10, 
                        'Soğan': 1, 
                        'Mısır': 1, 
                        }

#Classlarımızı oluşturalım
class CustomerToppings():
    """ Pizzanız için sos seçiniz."""
    
    def __init__(self, numToppings):
        self.numToppings = numToppings

    def set_toppings(self, numToppings):
        self.numToppings = numToppings

    def get_toppings(self):
        return topping_cost_mappings[self.numToppings]


class ToppingOrder():

    def __init__(self):
        self.topping = []

    def addTopping(self, toppings): 
        self.topping.append(toppings)

    def toppingTotal(self):
        get_topping_total = 0
        for toppings in self.topping:
            get_topping_total += toppings.get_toppings()
        return get_topping_total

topping_order = ToppingOrder()

#Kullanıcıdan ek malzemeleri seçmesini isteyelim ve son fiyatı hesaplayalım
def runTopping():
    print("\nPizzanız için ek ürün seçiniz.\n\n\
    _____________________________________________________\n\
    |   1: Zeytin   |    2: Mantar    | 3: Keçi Peyniri |\n\
    |     4: Et     |    5: Soğan     |     6: Mısır    |\n\
    |_______________|_________________|_________________|\n\
    Toplam sipariş miktarını görmek için 'f' tuşuna basınız: \n")

    while True:
        try: 
            response = input('-')
            if response == 'f':
                break
            toppings_wanted = int(response)

            toppings_wanted = topping_mappings[toppings_wanted]

            print(f"Ek ürün: {toppings_wanted}")
            topping_order.addTopping(CustomerToppings(toppings_wanted))
        except:
            print("Yanlış seçim yaptınız, lütfen tekrar deneyin.")

runTopping()

sub_description = int(order.getTotal())
sub_toppings =  int(topping_order.toppingTotal())
final_total = sub_description + sub_toppings

print(f" \nSipariş tutarınız ${final_total}\n")


#Pandas kullanarak dosyayı okuyalım 
veri= pd.read_csv("Orders_Database.csv")
veri



```
